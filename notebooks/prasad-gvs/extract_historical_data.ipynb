```python
import pandas as pd

import requests
import json

import glob
```

```python
def extract_data(ticker):
    if len(ticker)>0:
        apikey = "" # get your own keys add into your environment variables and modify the code here.
        api_url = "https://financialmodelingprep.com/api/v3/historical-market-capitalization/" + ticker + "?limit=1000&from=2021-01-01&to=2024-08-31&apikey=" + apikey
    
        # Make the API request
        response = requests.get(api_url)
    
        # Check if the request was successful (status code 200)
        if response.status_code == 200:
            response_data = response.json()
        
            # Save the JSON response to a file
            filename = ticker + ".json"
            with open(filename, 'w') as json_file:
                json.dump(response_data, json_file, indent=4)
        
            print(f"API response saved to " + filename)
        else:
            print(f"Failed to retrieve data from API. Status code: {response.status_code}")
    else:
        print(f"Invalid Ticker Symbol")
```

```python
def merge_json_to_csv(file_pattern):
    # Get a list of file paths matching the pattern
    json_files = glob.glob(file_pattern)

    # Load each CSV file into a DataFrame and store it in a list
    dfs = [pd.read_json(file) for file in json_files]

    # Concatenate all DataFrames into one
    combined_df = pd.concat(dfs, ignore_index=True)

    combined_df.to_csv('MarketCap_History_Top-100_Companies.csv',index=False)
```

```python
file_pattern = "extract_historical_data/*.json"
merge_json_to_csv(file_pattern)
```
